## 5 · `for` and `while` loops

**`for`** — repeat a known number of times; `range(5)` → 0..4.  
**`while`** — repeat until the condition is false; change something inside the loop so it can stop.

**Together:** write a **`while`** loop that keeps asking the user to guess a **secret word** until they type it correctly (pick the secret in a variable at the top).


```python
for i in range(5):
    print("range index", i)
```

```python
count = 0
while count < 5:
    print("count is", count)
    count += 1
```

```python
secret = "code"
guess = ""
while guess != secret:
    guess = input("Guess the secret word: ")
print("You got it!")
```
