## 6 · Manipulate DataFrames

Filter, add columns, sort, and summarize — same ideas as sorting a playlist or a snack list.


```python
df = pd.read_csv('week2_animals.csv')
forest_animals = df[df['habitat'] == 'forest']
df = df.copy()
df['heavy'] = df['weight_kg'] >= 50
by_weight = df.sort_values('weight_kg', ascending=False)
avg_weight_by_diet = df.groupby('diet')['weight_kg'].mean().round(1)
print(forest_animals[['name', 'habitat']])
print(by_weight[['name', 'weight_kg']].head(3))
print(avg_weight_by_diet)
```

## 7 · DataFrame cheatsheet (runnable)

Run each cell below. Uses a fresh copy of `week2_animals.csv`.

Answer key: lesson page → **Resources** tab.


```python
# Setup for cheatsheet demos
df_cs = pd.read_csv('week2_animals.csv')
print('Rows, columns:', df_cs.shape)
```

### Load & inspect


```python
df_demo = pd.read_csv('week2_animals.csv')
print('--- head(3) ---')
print(df_demo.head(3))
print('shape:', df_demo.shape)
print('columns:', df_demo.columns.tolist())
print('--- describe() ---')
print(df_demo.describe())
```

### Select rows & columns


```python
print(df_cs['name'])
print(df_cs[['name', 'habitat']])
print('habitat of row 0:', df_cs.loc[0, 'habitat'])
print('diet of row 0 (by position):', df_cs.iloc[0, 2])
```

### Filter & sort


```python
print(df_cs[df_cs['weight_kg'] >= 50][['name', 'weight_kg']])
print(df_cs.query('is_nocturnal == True')[['name', 'is_nocturnal']])
print(df_cs.sort_values('weight_kg', ascending=False)[['name', 'weight_kg']])
print(df_cs[df_cs['habitat'].isin(['forest', 'ocean'])][['name', 'habitat']])
```

### Add, change, drop


```python
work = df_cs.copy()
work['food_len'] = work['favorite_food'].str.len()
print(work[['name', 'favorite_food', 'food_len']].head(3))
print(work.assign(big=work['weight_kg'] >= 50)[['name', 'big']].head(3))
print(work.rename(columns={'name': 'animal'}).columns.tolist())
```

### Aggregate & count


```python
print(df_cs.groupby('habitat')['weight_kg'].mean().round(1))
print(df_cs['weight_kg'].sum())
print(df_cs['diet'].value_counts())
print(df_cs.isna().sum())
```
