## 3 · Tuples — practice


```python
snacks = (('chips', 7), ('apple', 9), ('cookie', 10))
total = 0
for snack, rating in snacks:
    total += rating
print('Average rating:', total / len(snacks))
```

## 4 · Dictionaries — practice


```python
creature = {
    'name': 'Mochi',
    'species': 'cat',
    'sounds': ['meow', 'purr'],
}
age = creature.get('age', 'unknown age')
print(f"{creature['name']} the {creature['species']} goes {creature['sounds'][0]} ({age}).")
```

## 6 · Manipulate — practice


```python
df = pd.read_csv('week2_animals.csv')
night = df[df['is_nocturnal'] == True]
print(night[['name', 'is_nocturnal']])
df = df.copy()
df['food_len'] = df['favorite_food'].str.len()
top3 = df.sort_values('food_len', ascending=False).head(3)
print(top3['name'].tolist())
```

## 8 · Challenge — animal data


```python
df = pd.read_csv('week2_animals.csv')

print('Avg weight by diet:')
print(df.groupby('diet')['weight_kg'].mean().round(1))

forest_night = df[(df['habitat'] == 'forest') & (df['is_nocturnal'] == True)]
print('Forest + nocturnal:', forest_night['name'].tolist())

df = df.copy()
df['size'] = df['weight_kg'].apply(lambda w: 'big' if w >= 50 else 'small')
print(df[['name', 'weight_kg', 'size']])

print('Top favorite_food:', df['favorite_food'].value_counts().idxmax())
```
