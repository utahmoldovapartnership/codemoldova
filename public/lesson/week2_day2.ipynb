## 4 · HTTP status codes cheatsheet

| Code | Meaning | When you see it |
|------|---------|-----------------|
| **200** | OK | Request worked. Data is in the response. |
| **404** | Not found | Bad URL or missing resource |
| **429** | Too many requests | Slow down and retry later |
| **500** | Server error | Their server broke — not your fault |

**Rule:** `2xx` = success · `4xx` = your problem · `5xx` = their problem

Also check `ok` in the JSON body. Empty search may return `ok: false`.


## 5 · Search your first movie


```python
import requests

query = "inception"
url = f"https://imdb.iamidiotareyoutoo.com/search?q={query}"
r = requests.get(url, timeout=10)
print("status:", r.status_code)
data = r.json()
print("ok:", data.get("ok"))
for movie in data.get("description", [])[:3]:
    print(movie["#TITLE"], movie["#YEAR"], "rank", movie["#RANK"])
    print("  cast:", movie["#ACTORS"])
```

## 6 · Browse a full film catalog

Studio Ghibli API — one list of every film. Great for loops and comparing `rt_score`.


```python
import requests

r = requests.get("https://ghibliapi.vercel.app/films", timeout=10)
print("status:", r.status_code)
films = r.json()
print(f"{len(films)} films")
for film in films[:5]:
    print(film["title"], "—", film["director"], f'{film["running_time"]} min, score {film["rt_score"]}')
```

## 7 · Compare three searches


```python
import requests

searches = ["inception", "matrix", "spirited away"]
for title in searches:
    r = requests.get(f"https://imdb.iamidiotareyoutoo.com/search?q={title}", timeout=10)
    if r.status_code == 200 and r.json().get("ok"):
        top = r.json()["description"][0]
        print(f"{title!r} → {top['#TITLE']} ({top['#YEAR']}) rank {top['#RANK']}")
```

## 8 · When the API says no


```python
import requests

def search_movies(query):
    url = f"https://imdb.iamidiotareyoutoo.com/search?q={query}"
    try:
        r = requests.get(url, timeout=10)
    except requests.RequestException as e:
        print("Network error:", e)
        return []
    if r.status_code != 200:
        print(f"{query}: status {r.status_code}")
        return []
    data = r.json()
    if not data.get("ok"):
        print(f"{query}: API error")
        return []
    return data.get("description", [])

hits = search_movies("inception")
print(hits[0]["#TITLE"] if hits else "no results")
print(search_movies(""))
```

## 9 · Save a movie snapshot


```python
from pathlib import Path
import json
import requests

Path("data").mkdir(exist_ok=True)
r = requests.get("https://imdb.iamidiotareyoutoo.com/search?q=inception", timeout=10)
data = r.json()
if data.get("ok") and data.get("description"):
    Path("data/inception.json").write_text(json.dumps(data["description"][0], indent=2))
    print("saved data/inception.json")
```
