## 4 · HTTP status codes cheatsheet

Every response has a **status code**. You print it with `r.status_code`.

| Code | Meaning | When you see it |
|------|---------|-----------------|
| **200** | OK | Request worked. Data is in the response. |
| **404** | Not found | Country or URL does not exist |
| **429** | Too many requests | Slow down and retry later |
| **500** | Server error | Their server broke — not your fault |

**Rule:** `2xx` = success · `4xx` = your problem · `5xx` = their problem

A fake country like `atlantis` returns **404**. Moldova returns **200**.


## 5 · Fetch Moldova’s passport

REST Countries returns a **list** — we use `[0]` for the first match.


```python
import requests

FIELDS = "name,capital,population,languages,borders,flags"
url = f"https://restcountries.com/v3.1/name/moldova?fields={FIELDS}"
r = requests.get(url, timeout=10)
print("status:", r.status_code)
moldova = r.json()[0]
print(moldova["name"]["common"])
print("capital:", moldova["capital"][0])
print("population:", f'{moldova["population"]:,}')
print("languages:", list(moldova["languages"].values()))
print("borders:", moldova["borders"])
```

## 6 · Loop over neighbors

Border codes: **ROU** = Romania, **UKR** = Ukraine. Fetch each with the `alpha` endpoint.


```python
import requests

FIELDS = "name,population,capital"
for code in ["ROU", "UKR"]:
    r = requests.get(f"https://restcountries.com/v3.1/alpha/{code}?fields={FIELDS}", timeout=10)
    if r.status_code == 200:
        c = r.json()
        print(c["name"]["common"], "—", f'{c["population"]:,}', "people, capital", c["capital"][0])
```

## 7 · Compare three countries


```python
import requests

for country in ["moldova", "romania", "ukraine"]:
    r = requests.get(f"https://restcountries.com/v3.1/name/{country}?fields=name,population", timeout=10)
    if r.status_code == 200:
        d = r.json()[0]
        print(f"{d['name']['common']}: {d['population']:,}")
```

## 8 · When the API says no


```python
import requests

def fetch_country(name):
    url = f"https://restcountries.com/v3.1/name/{name}?fields=name,population"
    try:
        r = requests.get(url, timeout=10)
    except requests.RequestException as e:
        print("Network error:", e)
        return None
    if r.status_code != 200:
        print(f"{name}: status {r.status_code}")
        return None
    return r.json()[0]

print(fetch_country("moldova")["name"]["common"])
print(fetch_country("atlantis"))
```
