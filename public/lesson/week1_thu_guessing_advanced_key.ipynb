## 1 · Core advanced (random, attempts, praise)


```python
import random


def check_guess(secret, guess, attempt_num):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print(f"You got it in {attempt_num} attempts!")


secret_number = random.randint(1, 100)
guess = 0
attempts = 0

print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 100.")

while guess != secret_number:
    guess = int(input("Your guess: "))
    attempts += 1
    check_guess(secret_number, guess, attempts)

print(f"Your final score: {attempts} attempts.")

if attempts <= 5:
    print("Incredible! You're a mind reader.")
elif attempts <= 10:
    print("Solid effort.")
else:
    print("Keep practicing!")
```

## 2 · Go beyond — play again with fresh random


```python
import random


def check_guess(secret, guess, attempt_num):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print(f"You got it in {attempt_num} attempts!")


def play_round(low, high):
    secret = random.randint(low, high)
    guess = 0
    attempts = 0
    print(f"I'm thinking of {low}–{high}.")
    while guess != secret:
        guess = int(input("Your guess: "))
        attempts += 1
        check_guess(secret, guess, attempts)
    return attempts


while True:
    play_round(1, 100)
    if input("Another game? (y/n): ").strip().lower() != "y":
        break
print("Bye!")
```

## 4 · Go beyond — Easy / Medium / Hard menu


```python
import random


def check_guess(secret, guess, attempt_num):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print(f"You got it in {attempt_num} attempts!")


print("Pick difficulty: 1 = Easy (1–10), 2 = Medium (1–50), 3 = Hard (1–100)")
choice = input("Enter 1, 2, or 3: ").strip()
if choice == "3":
    lo, hi = 1, 100
elif choice == "2":
    lo, hi = 1, 50
else:
    lo, hi = 1, 10

secret_number = random.randint(lo, hi)
guess = lo - 1
attempts = 0
print(f"Guess from {lo} to {hi}.")

while guess != secret_number:
    guess = int(input("Your guess: "))
    attempts += 1
    check_guess(secret_number, guess, attempts)
```

## 5 · Optional — safe int input


```python
def read_int(prompt):
    while True:
        raw = input(prompt)
        try:
            return int(raw)
        except ValueError:
            print("Please type a whole number.")


# Example: n = read_int("Your guess: ")
print("Use read_int() instead of int(input()) where you ask for guesses.")
```
