## 7 · Multi-turn chat (the agent)

`client.chats.create` keeps conversation history for you.

Type `quit` to exit.


````python
from google import genai

client = genai.Client()

SYSTEM = "You are a patient Python tutor for beginners. Keep answers under 80 words."

chat = client.chats.create(
    model="gemini-2.0-flash",
    config={"system_instruction": SYSTEM},
)

print("Chat agent ready. Type quit to exit.\n")

while True:
    user_text = input("You: ").strip()
    if user_text.lower() in ("quit", "exit", "q"):
        print("Bye!")
        break
    if not user_text:
        print("(empty message — type something)")
        continue
    # send to Gemini and get reply
    response = chat.send_message(user_text)
    print(f"\nBot: {response.text}\n")
````
