## H1 · Temperature (hot / warm / cold)

Rules from class: **hot** if temperature is greater than 30, **warm** if between 15 and 30 (inclusive), **cold** if less than 15.


```python
t = float(input("Temperature in °C? "))
if t > 30:
    print("hot")
elif t >= 15:
    print("warm")
else:
    print("cold")

# Example test notes (run with input or temporarily assign t):
# t = 10  → cold
# t = 20  → warm
# t = 35  → hot
```

## H4 · double(n) and five calls
