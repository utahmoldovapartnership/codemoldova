## 1 · Core game (fixed secret)


```python
def check_guess(secret, guess):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("You got it!")


secret_number = 7
guess = 0

while guess != secret_number:
    guess = int(input("Guess a number between 1 and 10: "))
    check_guess(secret_number, guess)

print("Game over!")
```

## 2 · Go beyond — welcome line


```python
def check_guess(secret, guess):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("You got it!")


secret_number = 7
guess = 0

print("Guess my number (1–10). You can do it.")

while guess != secret_number:
    guess = int(input("Your guess: "))
    check_guess(secret_number, guess)

print("Game over!")
```

## 3 · Go beyond — play again


```python
def check_guess(secret, guess):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("You got it!")


def play_one_round():
    secret_number = 7
    guess = 0
    while guess != secret_number:
        guess = int(input("Guess 1–10: "))
        check_guess(secret_number, guess)
    print("Round finished.")


while True:
    play_one_round()
    again = input("Play again? (y/n): ").strip().lower()
    if again != "y":
        print("Thanks for playing!")
        break
```

## 4 · Go beyond — max guesses (7)


```python
def check_guess(secret, guess):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("You got it!")


secret_number = 7
max_guesses = 7
used = 0
won = False

while used < max_guesses and not won:
    guess = int(input(f"Guess ({used}/{max_guesses} used, 1-10): "))
    used += 1
    check_guess(secret_number, guess)
    if guess == secret_number:
        won = True

if not won:
    print("Out of guesses. The number was", secret_number)

print("Game over!")
```

## 5 · Go beyond — Easy / Medium / Hard menu


```python
import random


def check_guess(secret, guess):
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("You got it!")


print("Pick difficulty: 1 = Easy (1–10), 2 = Medium (1–50), 3 = Hard (1–100)")
choice = input("Enter 1, 2, or 3: ").strip()
if choice == "3":
    lo, hi = 1, 100
elif choice == "2":
    lo, hi = 1, 50
else:
    lo, hi = 1, 10

secret_number = random.randint(lo, hi)
guess = lo - 1
print(f"Guess from {lo} to {hi}.")

while guess != secret_number:
    guess = int(input("Your guess: "))
    check_guess(secret_number, guess)

print("Game over!")
```
